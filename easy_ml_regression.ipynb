```python
#Importing pandas and the easy_ml library
import pandas as pd
from easy_ml import *

#Importing testing dataset
df = pd.read_csv('Concrete_Data.csv')
df.head()
```

```python
#Setting input features and target
features = df[['cement', 'slag', 'flyash', 'water', 
                'superplasticizer', 'coarseaggregate', 
                'fineaggregate', 'age']]

target = df[['csMPa']]

#Calling the easy_ml class with the StandardScaler applied
easy_ml = easy_ml(features, target, standard_scaler = True)
easy_ml.regression()
```
